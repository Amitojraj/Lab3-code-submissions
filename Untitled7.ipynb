```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Assuming you have a DataFrame named 'df' with columns 'Polar Surface Area' and 'measured log solubility in mols per litre'
X_column_name = ''
Y_column_name = 'measured log solubility in mols per litre'

# Extract the data from the specified columns
X = df[X_column_name].values.reshape(-1, 1)  # Reshape to a 2D array (required by scikit-learn)
Y = df[Y_column_name]

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, Y)

# Calculate the correlation coefficient and the p-value
correlation_coefficient, p_value = pearsonr(X.flatten(), Y)

print("Correlation Coefficient:", correlation_coefficient)
print("P-Value:", p_value)

# Make predictions with the model
Y_pred = model.predict(X)

# Plot the original data and the regression line
plt.scatter(X, Y, label='Data', color='blue')
plt.plot(X, Y_pred, label='Linear Regression', color='red')
plt.xlabel(X_column_name)
plt.ylabel(Y_column_name)
plt.legend()
plt.show()

# Get the regression coefficients
slope = model.coef_[0]
intercept = model.intercept_

print("Slope (Coefficient):", slope)
print("Intercept:", intercept)
```

```python
import requests

# URL of the raw CSV file on GitHub
url = "https://raw.githubusercontent.com/GLambard/Molecules_Dataset_Collection/master/latest/ESOL_delaney-processed.csv"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (HTTP status code 200)
if response.status_code == 200:
    # Specify the local file path where you want to save the downloaded file
    local_file_path = "downloaded_file.csv"

    # Write the content of the response to the local file
    with open(local_file_path, 'wb') as file:
        file.write(response.content)

    print(f"File '{local_file_path}' has been downloaded.")
else:
    print(f"Failed to download the file. Status code: {response.status_code}")
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
X=np.array([1,2,3,4,5])
Y=np.array([1.2,2.3,4,4.3,5.6])
model=LinearRegression()
X=X.reshape(-1,1)
model.fit(X,Y)

correlation_coefficient, p_value = pearsonr(X.flatten(), Y)



# Make predictions with the model
Y_pred = model.predict(X)

# Plot the original data and the regression line
plt.scatter(X, Y, label='Data', color='blue')
plt.plot(X, Y_pred, label='Linear Regression', color='red')
plt.xlabel('X_column_name')
plt.ylabel('Y_column_name')
plt.grid()
plt.legend()
plt.show()

# Get the regression coefficients
slope = model.coef_[0]
intercept = model.intercept_
print("Correlation Coefficient:", correlation_coefficient)
print("P-Value:", p_value)
print("Slope (Coefficient):", slope)
print("Intercept:", intercept)
```
